```python
# 2050. Parallel Courses III
from typing import List

class Solution:
    def minimumTime(self, 
                    n: int, 
                    relations: List[List[int]], 
                    time: List[int]) -> int:

        rel = {}  
        n_dep = [0] * n
        for link in relations:
            c1, c2 = link[0]-1, link[1]-1
            n_dep[c2] +=1
            if c1 not in rel:
                rel[c1] = [c2]
            else:
                rel[c1].append(c2)
        
        queue = [i for i, val in enumerate(n_dep) if val == 0 ]
        max_time = 0
        del_time = [0] * n
        while queue:
            i = queue.pop(0)
            max_time = max(max_time, time[i] + del_time[i])
            for j in rel.get(i, []):
                n_dep[j] = n_dep[j]-1
                del_time[j] = max(del_time[j], time[i] + del_time[i])
                if n_dep[j]==0:
                    queue.append(j)
                    
        return max_time
    
sol = Solution()

n = 3        
relations = [[1,3],[2,3]]
time = [3,2,5]
    

aa = sol.minimumTime(n, relations, time)
print (aa, aa==8)

n = 5
relations = [[1,5],[2,5],[3,5],[3,4],[4,5]]
time = [1,2,3,4,5]
aa = sol.minimumTime(n, relations, time)
print (aa, aa==12)

n = 9
relations = [[3,2],[3,1],[1,7],[2,7],[4,6],[2,9],[3,9],[4,9],[6,9],[8,9]]
time = [3,5,7,1,8,2,5,7,4]
aa = sol.minimumTime(n, relations, time)
print (aa, aa==17)    
    
```

```python


class Solution:
    def minimumTime(self, 
                    n: int, 
                    relations: List[List[int]], 
                    time: List[int]) -> int:
        
        n_rel = len(relations)
        idx_rel = list(range(n_rel))
        idx_rel.sort(key = lambda i: relations[i][0])
        
        rel = {}  
        
        n_dep = [0] * (n+1)        
        beg = [n_rel] * (n+1)
        end = [-1] * (n+1)
        for i in range(n_rel):
            link = relations[idx_rel[i]]
            c1, c2 = link[0], link[1]
            beg[c1] = min(beg[c1], i)
            end[c1] = max(end[c1], i)
            n_dep[c2] +=1
        
        queue = [i for i, val in enumerate(n_dep) if val == 0 ]
        max_time = 0
        del_time = [0] * (n+1)

        while queue:
            i = queue.pop()
            max_time = max(max_time, time[i-1] + del_time[i])

            for j in range(beg[i], end[i]+1):
  
                c1, c2 = relations[idx_rel[j]]
                
                n_dep[c2] -= 1
                del_time[c2] = max(del_time[c2], time[i-1] + del_time[i])
                if n_dep[c2]==0:
                    queue.append(c2)                   

                    
        return max_time

sol = Solution()

n = 3        
relations = [[1,3],[2,3]]
time = [3,2,5]
    

aa = sol.minimumTime(n, relations, time)
print (aa, aa==8)

n = 5
relations = [[1,5],[2,5],[3,5],[3,4],[4,5]]
time = [1,2,3,4,5]
aa = sol.minimumTime(n, relations, time)
print (aa, aa==12)

n = 9
relations = [[3,2],[3,1],[1,7],[2,7],[4,6],[2,9],[3,9],[4,9],[6,9],[8,9]]
time = [3,5,7,1,8,2,5,7,4]
aa = sol.minimumTime(n, relations, time)
print (aa, aa==17)
```

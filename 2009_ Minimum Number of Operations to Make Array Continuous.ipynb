```python
# 2009. Minimum Number of Operations to Make Array Continuous
from typing import List

class Solution:
    def minOperations(self, nums: List[int]) -> int:
        nums.sort()
        
        n = len(nums)
        i1 = 0
        i2 = 0

        best_penalty = n - 1
        dupl = 0
        for i1 in range(n):
            
            if i1>0 and nums[i1] == nums[i1-1]:
                dupl -=1                
                continue
                            
            while i2<i1:
                i2 +=1
                if nums[i2] == nums[i2-1]:
                    dupl +=1
                    
            i2 +=1
            if i2>=n:
                break
                
            if nums[i2] == nums[i2-1]:
                dupl +=1
                        
            while nums[i2] - nums[i1]<=n-1: 
                
                penalty = i1 + (n-(i2+1)) + dupl
                if penalty < best_penalty:
                    best_penalty = penalty
                    
                i2 +=1
                if i2>=n:
                    break
                
                if nums[i2] == nums[i2-1]:
                    dupl +=1
        
        return best_penalty
    
# nums = [10, 1000, 2000,3000]  # 3
# nums = [4,4,4,2,5,3] # 2
# nums = [1,2,3,5,6] # 1
# nums = [4,2,5,3] # 0
# nums = [2, 2, 4, 6, 9, 11, 12, 17, 23, 24, 25, 26, 30, 32, 40, 41, 44, 47, 49, 49, 49,54,51] # 13
# nums = [2, 4, 6, 9, 11, 12, 17, 23, 25, 26, 30, 32, 40, 41, 44, 47, 49, 49, 49] # 12 

# nums = [3016,3228,3079,2920,3059,3369,3221,3140,3236,3198,3332,3189,3075,3147,3058,3141,2970,3058,3164,2976,3190,2915,3060,3384,3195,2923,3172,2947,3127,3360,3001,3024,3088,3350,3313,2996,3279,3274,3017,3044,3286,3179,2958,3137,3165,3172,2932,3346,2919,3339,3313,3222,3226,3083,3264,2995,3064,3325,2924,3001,3013,3074,3188,3384,3296,2917,3259,3120,3330,3272,3155,2934,3382,2910,3220,3015,2949,3006,3313,3304,3229,3300,3288,3121,2915,3253,3043,3268,3376,2946,3193,3099,3391,3350,2984,3233,2946,3310,3081,3077,3119,3081,3106,3118,3081,3156,3230,3061,3019,3368,2927,3188,3354,3008,3270,3177,2926,3198,2923,3390,2919,3176,3317,3191,3114,3340,3365,2984,3170,2954,3272,3182,3249,3229,3079,3176,3236,3379,2913,3116,3167,3097,3103,3122,3295,3234,3244,2915,3353,3113,3128,3225,3010,3278,3048,3112,2951,2950,3220,3261,3332,2957,3024,3146,2994,3309,3223,3287,3026,3256,3158,3150,3205,3111,3275,3231,3256,3334,3289,3177,3362,3061,3145,3163,3123,3073,3336,3235,3259,3337,3356,3161,3103,2997,3388,2982,3245,2987,2973,3370,2999,2923,3028,3080,3337,2981,3295,2927,3082,3195,2963,3129,3274,2980,3084,3054,2959,3305,3173,3019,2936,3092,3362,3167,3097,3077,3166,3232,2965,3390,3104,3266,2913,2984,3311,3031,3235,3120,3083,2943,3294,3306,3079,3292,3253,3049,3391,3354,3129,3088,2957,2958,2926,3343,3126,3069,3183,3364,2913,3353,3299,3096,2923,3299,3243,3154,2921,3129,3014,3379,2989,2957,3038,3390,3225,3337,3360,2913,3251,3200,3080,3171,3118,3230,3320,2998,3354,2934,3294,3105,2926,3295,3116,3294,3089,3316,3079,3152,3186,3012,3106,3056,3140,3265,3061,3292,3062,3254,2987,3125,3303,3052,3154,3249,3359,3297,3037,3147,2925,3160,3148,3321,3220,2958,3107,3233,3059,3383,3264,2959,3200,3154,3372,2930,3080,2915,3274,3204,2933,3214,3058,3187,3274,3095,3247,3318,3174,3149,3097,2990,3353,3001,3150,3161,3077,3148,3061,3016,3291,3299,3313,3115,3126,3176,3021,3356,3042,3110,2921,2920,3386,3209,3039,3371,3079,3128,3265,2982,3199,3110,3133,3258,3206,2924,3169,3245,3331,3338,3288,3205,3051,3264,3032,3206,3268,2970,3050,2983,3080,2934,3199,3273,3335,2938,3355,3067,3119,3229,2977,3094,3271,3294,3029,3332,3375,3298,3261,3328,3019,3173,3274,3169,3074,2939,3202,3368,3331,3285,2992,3307,3041,2962,3224,3066,3215,3062,3354,3389,3369,3315]
# 171

sol = Solution()
sol.minOperations(nums)
```

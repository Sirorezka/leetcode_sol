```python
# 1458. Max Dot Product of Two Subsequences
from typing import List

class Solution:
    def maxDotProduct(self, nums1: List[int], nums2: List[int]) -> int:
        
        n1, n2 = len(nums1), len(nums2)
        max_sol = [float('-inf')] * (n2 +1)
        
        for i in range(0, n1):
            prev_val = 0
            val = nums1.pop()
            for j in range(0, n2):
                v1, v2 = max_sol[j+1], max_sol[j]
                t = val * nums2[n2-1-j]
                max_sol[j+1] = max(v1, v2, t + max(0,prev_val))
                if i !=0:
                    prev_val = v1
                
        return max_sol[-1]
    
    
        
        
nums1 = [2, 1, -2, 5]
nums2 = [3, 0, -6]

sol = Solution()
aa = sol.maxDotProduct(nums1, nums2)
print (aa, aa == 18)

nums1 = [3,-2]
nums2 = [2,-6,7]
sol = Solution()
aa = sol.maxDotProduct(nums1, nums2)
print (aa, aa == 21)


nums1 = [-1,-1]
nums2 = [1,1]
sol = Solution()
aa = sol.maxDotProduct(nums1, nums2)
print (aa, aa == -1)

nums1 = [-5,-1,-2]
nums2 = [3,3,5,5]
sol = Solution()
aa = sol.maxDotProduct(nums1, nums2)
print (aa, aa == -3)
```

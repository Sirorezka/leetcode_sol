```python
# 34. Find First and Last Position of Element in Sorted Array

class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        self.n = len(nums)
        n = self.n
        
        if self.n==0:
            return -1, -1
        
        n1 = 0
        n2 = self.n - 1
        
        if nums[n1]>target or nums[n2]<target:
            return -1, -1
        
        while True:
            mid = int((n1+n2)/2) 
            if n2-n1<2:                
                break
            if nums[mid]<target:
                n1 = mid
            elif nums[mid]>target:
                n2 = mid
            else:
                break
          
        if n2-n1<2:                
            if nums[n1]==target:
                mid = n1
            elif nums[n2]==target:
                mid = n2
            else:
                return -1, -1
                
        if nums[n1]==target:
            low = n1
        else:
            low = mid
            while True:
                t = int((n1+low)/2)
                if nums[t] < target:
                    n1 = t
                else:
                    low = t
                if low-n1<2:
                    break
            
        if nums[n2]==target:
            high = n2
        else:
            high = mid
            while True:
                t = int((n2+high)/2)
                if nums[t] > target:
                    n2 = t
                else:
                    high = t
                if n2-high<2:
                    break
                    
        return low, high
    
# nums = [5,7,7,7,7,7,7,7,8,8,10] ## (8,9)
# target = 8

# nums = [8,8,8,8,8,8,8,8,10] ## (0,7)
# target = 8

# nums = [] ## (-1,-1)
# target = 0

# nums = [1,1,1,1,1] ## (0,4)
# target = 1

# # nums = [5,7,7,8,8,10]
# target=6

# nums = [1,4]
# target=4

sol = Solution()
aa = sol.searchRange(nums, target)
print (aa)
```
